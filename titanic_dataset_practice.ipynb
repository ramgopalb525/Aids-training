```python
import seaborn as sns
```

```python
colors=['pink','grey','black']
sns.barplot(data=train,x="Pclass",y="Survived",errorbar=None,palette=colors)
```

```python
sns.scatterplot(data=train,x="Age",y="Fare",hue='Survived',palette='viridis')
```

```python
pair_data=train[['Age','Fare','Pclass']]
sns.pairplot(data=pair_data,hue="Pclass",palette='coolwarm')
```

```python
train_heat=train.select_dtypes(include=['number'])
correlation_matrix=train_heat.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="cividis",fmt=".2f")
plt.show()
```
