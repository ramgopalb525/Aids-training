```python
import matplotlib.pyplot as plt


months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [1200, 1500, 1800, 1700, 1600, 2200, 2100, 2500, 2400, 2300, 3000, 2500]

plt.figure(figsize=(10, 6))
plt.plot(months, sales, marker='o', color='b', linestyle='-')
plt.title('Monthly Sales Tracking', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Sales Amount ' )
plt.grid(True)
plt.show()
```

```python

ages = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]
spending = [150, 180, 200, 210, 250, 230, 300, 320, 280, 350, 330, 400, 450]

plt.figure(figsize=(10, 6))
plt.scatter(ages, spending, color='b', edgecolor='black')
plt.title('Age Distribution vs Spending Pattern', )
plt.xlabel('Age', fontsize=12)
plt.ylabel('Spending ($)', fontsize=12)


```

```python
import matplotlib.pyplot as plt
import numpy as np


subjects = ['Math', 'Science', 'English']
marks = {
    'Math': [80, 75, 85, 90, 95, 60, 70, 100, 110, 45],  
    'Science': [88, 92, 95, 100, 105, 60, 85, 70, 78, 89],  
    'English': [70, 72, 68, 80, 85, 90, 95, 100, 60, 55]  
}


subject_means = [np.mean(marks[subject]) for subject in subjects]

plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1) 
plt.bar(subjects, subject_means, color=['blue', 'green', 'orange'])
plt.title('Average Marks Across Subjects', fontsize=14)
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Average Marks', fontsize=12)


plt.subplot(1, 2, 2)  
plt.boxplot([marks['Math'], marks['Science'], marks['English']], labels=subjects)
plt.title('Marks Distribution and Outliers')
plt.xlabel('Subjects')
plt.ylabel('Marks')



plt.show()
```
